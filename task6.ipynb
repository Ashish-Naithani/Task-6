```python
#Using database ''customerdb''
query='''USE customerdb;'''
cursor.execute(query)
cursor.fetchall()
```

```python
#Count of Distinct order id's in the dataset
query='''SELECT COUNT(DISTINCT order_id) AS Number_of_id
FROM olist_order_items_dataset;'''
cursor.execute(query)
cursor.fetchall()
```

```python
#Sum of price for Distinct order id's in the dataset
query='''SELECT DISTINCT order_id,SUM(price) AS Total_Price
FROM olist_order_items_dataset
GROUP BY order_id
ORDER BY Total_Price DESC;'''
cursor.execute(query)
cursor.fetchall()
```

```python
#Extracting Month from Date
query='''SELECT EXTRACT(MONTH FROM STR_TO_DATE(shipping_limit_date, '%d/%m/%Y %H:%i')) AS MONTH_OF_DATE
FROM olist_order_items_dataset
WHERE shipping_limit_date IS NOT NULL;
'''
cursor.execute(query)
cursor.fetchall()
```

```python
#Using Group by with Extracting Month from Date
query='''SELECT EXTRACT(YEAR FROM STR_TO_DATE(shipping_limit_date, '%d/%m/%Y %H:%i')) AS YEAR_OF_DATE,
SUM(price)
FROM olist_order_items_dataset
WHERE shipping_limit_date IS NOT NULL
GROUP BY  YEAR_OF_DATE;
'''
cursor.execute(query)
cursor.fetchall()
```

```python
#Using limit in Extracting Month from Date
query='''SELECT EXTRACT(YEAR FROM STR_TO_DATE(shipping_limit_date, '%d/%m/%Y %H:%i')) AS YEAR_OF_DATE,
SUM(price)
FROM olist_order_items_dataset
WHERE shipping_limit_date IS NOT NULL
GROUP BY  YEAR_OF_DATE
LIMIT 2;
'''
cursor.execute(query)
cursor.fetchall()
```
